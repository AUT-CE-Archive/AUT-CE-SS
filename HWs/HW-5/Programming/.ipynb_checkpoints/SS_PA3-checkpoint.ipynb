# Part 1

## a) Coefficients Calculator

```python
# Calculate b & c coefficients
def calc_fourier_coefs(data, k, T):
    w = (np.pi* 2 / T)
    _sum = []
    for i in range(k):
        b = sum([data[j] * np.cos(j * i * w) for j in range(T - 1)]) / T
        c = sum([data[j] * np.sin(j * i * w) for j in range(T - 1)]) / T
        _sum.append(complex(b, -c))
    return b, c, _sum
```

## Coefficients Plot

```python
def plot_fourier_coefs(_sum, k, t):

    # Calculate absolute
    _sum = [abs(tmp) for tmp in _sum]

    # X-axis
    freq = np.linspace(0, t, k)

    fig = plt.figure(figsize = (20, 5))
    plt.stem(freq, _sum)
    plt.show()
```

```python
# PLot CT
def plot_ct_signal(t, x_t):
    fig = plt.figure(figsize=(20, 7))
    plt.xticks(t)
    plt.plot(t, x_t)
    plt.show()
```

## Right Plot

```python
t_1 = np.arange(-6, -1, step = 1)
t_2 = np.arange(-2, 3, step = 1)
t_3 = np.arange(2, 7, step = 1)

t = np.concatenate([t_1, t_2, t_3])
def x_t(t, a):
    return t + a

# Vectorize and map
x1_t = np.vectorize(x_t, otypes = [float])(t_1, 4)
x2_t = np.vectorize(x_t, otypes = [float])(t_2, 0)
x3_t = np.vectorize(x_t, otypes = [float])(t_3, -4)
x_t = np.concatenate([x1_t, x2_t, x3_t])

plot_ct_signal(t, x_t)
```

## Read Data

```python
# Read sound file
sample_rate, data = wavfile.read('sound.wav')
length = data.shape[0] / sample_rate

print(f"number of channels = {data.shape}")
print(f"length = {length}s")
```

## a) Plot sound.wav

```python
sample_size = 1000

sampled_data = data[:sample_size]
t = np.arange(0, sample_size / sample_rate, 1 / sample_rate)

fig = plt.figure(figsize = (20, 5))
plt.plot(t, sampled_data)
plt.ylabel('sample data')
plt.xlabel('time [s]')
plt.show()
```

## b) Plot a_k & b_k

```python
T = 61   # Period (rounded)
k = 150  # Coefficients counts
_sum = []

# Calculate b & c, then save the complex form (From part )
b, c, _sum = calc_fourier_coefs(data, k, T)

# Plot the coefficients
plot_fourier_coefs(_sum, k, sample_rate)
```

## c) Frequency

```python
start_time, end_time = 0 , 1000

# Define start and end points
start_point = int(sample_rate * start_time / 1000)
end_point = int(sample_rate * end_time / 1000)

# Count freq total
counter = sum([1 for i in range(start_point, end_point) if (data[i] < 0 and data[i + 1] > 0)])

avg_freq = counter / length
print(f'Average Freq: {avg_freq}')
```
