## 1. Signal plotting
In this section, we learn how to plot signals by some examples to become familiar with some basic functions of **NumPy** and **Matplotlib**.

To define a signal using NumPy, you need to create two arrays $t$ and $x(t)$ in continuous-time case, or $n$ and $x[n]$ in discrete-time case.
For plotting the signals, we have defined plot_ct_signal (for continuous-time signals) and plot_dt_signal (for discrete-time signals). You have to write the body of these functions:

* `plot_ct_signal(t, x_t)`
* `plot_dt_signal(n, x_n)`


For example, to plot the following signals:

* Continuous-Time: $ x(t)\:=\:sin(t)\:\:;\:\:\:step\:=\:0.01\:\:;\:\:interval\:=\:[-3\pi \:,\:3\pi] $
* Discrete-Time: $ x[n]\:=\:sin(n)\:\:;\:\:interval\:=\:[-10 \:,\:10] $

We can use:
```python
t = np.arange(-3 * np.pi, 3 * np.pi, step=0.01)
x_t = np.sin(t)
plot_ct_signal(t=t, x_t=x_t)
```

Or for DT:

### 1.a) Complete the following functions.

````python
def plot_ct_signal(t, x_t):
    fig = plt.figure(figsize=(14.5, 7))
    plt.plot(t, x_t)
    plt.show()
````

````python
t = np.arange(-3 * np.pi, 3 * np.pi, step=0.01)
x_t = np.sin(t)
plot_ct_signal(t=t, x_t=x_t)
````

````python
def plot_dt_signal(n, x_n):
    fig = plt.figure(figsize=(14.5, 7))
    plt.stem(n, x_n)
    plt.show()
````

````python
n = np.arange(-10, 10, step=1)
x_n = np.sin(n)

plot_dt_signal(n=n, x_n=x_n)
````

### 1.b) Now plot the following signals.

#### Continuous-Time signals:

1. $x_1(t)=e^{2t} \: \cos(2t) \: u(-t+4) \; ; \quad \mathrm{interval}=[-10, 10]$

````python
t = np.arange(-10, 10, step=0.1)
x1_t = np.exp(2 * t) * np.cos(2 * t) * np.heaviside(-t + 4, 0)

plot_ct_signal(t = t, x_t = x1_t)
````

2. $x_2(t) = 
	\left\{\begin{matrix}
		 1 & t < -1 \\
		 -t & -1 \leq t \leq 1 \\
		 \sin(t-1) & t > 1 \\
	\end{matrix}\right. 
	\; ; \quad \mathrm{interval}=[-4, 4]$

````python
# x2_t = lambda x: 1 if (x < -1) else (-x if (-1 <= x <= 1) else np.sin(x - 1))

def x2_t(x):
    if x < -1:
        return 1
    elif -1 <= x <= 1:
        return -x
    else:
        return np.sin(x - 1)

t = np.arange(-4, 4, step = 0.01)

x2_t = 2 * np.vectorize(x2_t, otypes=[float])(t)  # Vectorize and map
plot_ct_signal(t = t, x_t = x2_t)
````

3. $x_3(t) = \sum_{n=-20}^{20} e^{-|2t+n|}
	\; ; \quad \mathrm{interval}=[-5, 5]$

````python
x3_t = lambda x: sum([np.exp(- abs((2 * x) + n)) for n in range(-20, 21)])
t = np.arange(-5, 5, step = 0.01)

x_t = 2 * np.vectorize(x3_t)(t)   # Vectorize and map
plot_ct_signal(t = t, x_t = x_t)
````

#### Discrete-Time signals:

1. $ x_1[n] = \cos(3n) \; ; \quad \mathrm{interval}=[-5, 5] $

````python
n = np.arange(-5, 5, step = 1)
x1_n = np.cos(3 * n)

plot_dt_signal(n = n, x_n = x1_n)
````

2. $ x_2[n] = \cos(3 \pi n) \; ; \quad \mathrm{interval}=[-5, 5] $

````python
n = np.arange(-5, 5, step = 1)
x2_n = np.cos(3 * n * np.pi)

plot_dt_signal(n = n, x_n = x2_n)
````

3. $ x_3[n]=\sin(0.2\pi n)+\cos(2.2\pi n) \; ; \quad \mathrm{interval}=[-20, 20] $

````python
n = np.arange(-20, 20, step = 1)
x3_n = np.sin(0.2 * np.pi * n) + np.cos(2.2 * np.pi * n)

plot_dt_signal(n = n, x_n = x3_n)
````

4. $ x_4[n]=\sin(2.2\pi n)+\cos(4.2\pi n) \; ; \quad \mathrm{interval}=[-20, 20] $

````python
n = np.arange(-20, 20, step = 1)
x4_n = np.sin(2.2 * np.pi * n) + np.cos(4.2 * np.pi * n)

plot_dt_signal(n = n, x_n = x4_n)
````

5. $ x_5[n] = u[n+3]-u[-n+3]+2\delta[n]
	\; ; \quad \mathrm{interval}=[-10, 10] $

````python
n = np.arange(-10, 10, step = 1)
unit_pulse_arr = [0 if (i != 0) else 2 for i in range(-10, 10)]
x4_n = np.heaviside(n + 3, 0) - np.heaviside(-n + 3, 0) + unit_pulse_arr

plot_dt_signal(n = n, x_n = x4_n)
````

# !!! Results of the Analysis !!!
- System is ***Time-Invarient***. This result was acquired by testing the two funtions `sin(n)` adn `sin(n + 2)`. The experiment was done by the code below:

Original funtion:
```
n = np.arange(-10, 10, step=1)
x_n = np.sin(n)
n1, y_n1 = get_system_output(n, x_n)
plot_dt_signal(n=n1, x_n=y_n1)
```
Time-shifted funtion:
```
n = np.arange(-10, 10, step=1)
x_n = np.sin(n + (2 * np.pi))
n1, y_n1 = get_system_output(n, x_n)
plot_dt_signal(n=n1, x_n=y_n1)
```

- The system is also ***Linear***. We know when ever the system is time-scaling, it will also be Linear. In other words, system linearity is independent of time-scaling.
